Black Scholes Model and results are correct

```python
%%writefile bs.m

S = 100
X = 100
r = 0.05
v = 0.2
T = 1

%Black and Scholes in Matlab

d1=(log(S/X)+(r+v^2/2)*T)/(v*sqrt(T));
d2=d1-v*sqrt(T);
c = S*normcdf(d1)-X*exp(-r*T)*normcdf(d2)


```
